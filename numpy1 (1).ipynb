```python
zeros_array = np.zeros((3, 4))
print("\n3x4 zeros array:\n", zeros_array)
```

```python
ones_array = np.ones((2, 3, 4), dtype=np.int16)
print("\n2x3x4 ones array (int16):\n", ones_array)
```

```python
range_array = np.arange(10, 25, 5)
print("\nValues from 10 to 25 with step 5:\n", range_array)
```

```python
linspace_array = np.linspace(0, 2, 9)
print("\n9 evenly spaced values between 0 and 2:\n", linspace_array)
```

```python
constant_array = np.full((2, 2), 7)
print("\n2x2 constant array (7):\n", constant_array)
```

```python
identity_matrix = np.eye(2)
random_matrix = np.random.random((2, 2))
```

```python
print("\n2x2 Identity Matrix:\n", identity_matrix)
print("\n2x2 Random Matrix:\n", random_matrix)
```

```python
for name, arr in zip(['a', 'b', 'c'], [a, b, c]):
    print(f"\nArray {name}:")
    print("Shape:", arr.shape)
    print("Dimensions (ndim):", arr.ndim)
    print("Size:", arr.size)
```

```python
print("\nArray b dtype:", b.dtype)
print("Array b dtype name:", b.dtype.name)
```

```python
b_int = b.astype(int)
print("Array b converted to integers:\n", b_int)
```

```python
print("\na + b:\n", a + b)
print("\na - b:\n", a - b)
```

```python
print("\nb * b:\n", b * b)
print("\nb / b:\n", b / b)
```

```python
print("\nSquare root of b:\n", np.sqrt(b))
```

```python
print("\nSum:", np.sum(b))
print("Minimum:", np.min(b))
print("Maximum:", np.max(b))
print("Cumulative Sum:", np.cumsum(b))
print("Mean:", np.mean(b))
print("Median:", np.median(b))
print("Standard Deviation:", np.std(b))
```
